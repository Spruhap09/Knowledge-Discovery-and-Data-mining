```python
print('BTW, there is a typo - Smoker is Somker, I am not changing it')
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import CategoricalNB
```

```python
print('Loading the bp dataset')
bp = pd.read_csv('/Users/spruhap/Desktop/School/Fall23/CS513/midterm/BP.csv')
bp.head(10)
```

```python
print('Replacing the string values for Smoker and Gender with int values')
bp['Somker'] = bp['Somker'].replace('1- Non-smoker', 1)
bp['Somker'] = bp['Somker'].replace('2- Light', 2)
bp['Somker'] = bp['Somker'].replace('3- Moderate', 3)
bp['Somker'] = bp['Somker'].replace('4- Heavy', 4)
bp['Somker'] = bp['Somker'].replace('5- Very Heavy', 5)
bp['Gender'] = bp['Gender'].replace('Female', 0)
bp['Gender'] = bp['Gender'].replace('Male', 1)
```

```python
print('Creating a data frame called target for the BP_Status variable')
target = bp['BP_Status'] # target variable
target.head(10)
```

```python
print('Dropping the BP_Status field')
# Splitting the dataset
attributes = bp.drop(columns = ['BP_Status'])
attributes.head(10)
```

```python
print('Dividing the data into test and train data given the 30/70 split ratio')
#RANDOM_STATE helps to keep the data consistent -> it keeps the values consistent when you run the code again
attribute_train, attribute_test, target_train,target_test  = train_test_split(attributes, target,test_size = 0.3, random_state = 87, shuffle = True)
```

```python
cart_model = DecisionTreeClassifier()
# given the train split, this fits and trains the data given your DecisionTreeClassifier
cart_model = cart_model.fit(attribute_train, target_train)
# AND then it predicts the value given our test data
target_pred = cart_model.predict(attribute_test);
# then we just print out the accuracy scores and other details
print("Accuracy:",accuracy_score(target_test, target_pred)) 
print("Confusion matrix:", confusion_matrix(target_test, target_pred))
print("Classification report:", classification_report(target_test, target_pred))
```

```python
plt.figure(figsize=(12, 6))
# this gives us the decision tree, the the target values includes in the class names because they are the only two options
plot_tree(cart_model, feature_names=attributes.columns, class_names=['High', 'Normal'], filled=True)
plt.title("CART Decision Tree")
plt.show()
```
