```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print('BTW, there is a typo - Smoker is Somker, I am not changing it')
bp = pd.read_csv('/Users/spruhap/Desktop/School/Fall23/CS513/midterm/BP.csv')
bp.head(10)
```

```python
print('The box plot for Age')
sns.boxplot(data=bp, y='Age')
plt.title('Box Plot for Age')
plt.show()
```

```python
print('The box plot for Height')
sns.boxplot(data=bp, y='Height')
plt.title('Box Plot for Height')
plt.show()
```

```python
print('The box plot for Weight')
sns.boxplot(data=bp, y='Weight')
plt.title('Box Plot for Weight')
plt.show()
```

```python
print('Printing the pair plot for Age, Height, Weight')
sns.pairplot(bp[['Age', 'Height', 'Weight']])
plt.suptitle('Pair Plot for Age, Height, and Weight')
plt.show()
```
