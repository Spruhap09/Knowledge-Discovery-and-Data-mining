```python
print('BTW, there is a typo - Smoker is Somker, I am not changing it')
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
print('Loading the bp dataset')
bp = pd.read_csv('/Users/spruhap/Desktop/School/Fall23/CS513/midterm/BP.csv')
bp.head(10)
```

```python
print('Replacing the string values for Smoker and Gender with int values')
bp = bp.astype('category')
bp['Somker'] = bp['Somker'].cat.codes
bp['Gender'] = bp['Gender'].map({'Male': 1, 'Female': 0})
```

```python
print('Creating a data frame called target for the BP_Status variable')
target = bp['BP_Status'] # target variable
target.head(10)
```

```python
print('Dropping the BP_Status field')
# Splitting the dataset
attributes = bp.drop('BP_Status', axis=1)
attributes.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize MinMaxScaler
scaler = MinMaxScaler()
# Fit and transform the data - because the data is from diff units of mesurements 
attributes = pd.DataFrame(scaler.fit_transform(attributes), columns=attributes.columns)
attributes.head()
```

```python
print('Dividing the data into test and train data given the 30/70 split ratio')
#RANDOM_STATE helps to keep the data consistent -> it keeps the values consistent when you run the code again
attribute_train, attribute_test, target_train,target_test  = train_test_split(attributes, target,test_size = 0.3, random_state = 87, shuffle = True)
```

```python
print('Assuming k=2 based on problem 3')
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print('Doing the classification for k = 2')
print()
k = 2
#creates a default model with the no of neighbors you want
knn = KNeighborsClassifier(n_neighbors = k)

#this fits the data -> actually runs the classifier and fits the data actually
knn.fit(attribute_train, target_train)

#prediciting given the values from the non target variables
target_pred = knn.predict(attribute_test)

cm = confusion_matrix(target_test, target_pred)
print('Confusion Matrix for k = 2')
print(confusion_matrix(target_test, target_pred))
print()

print('Accuracy score for k = 2')
print(accuracy_score(target_test, target_pred))
print()

print('Classification Report for k = 2')
print(classification_report(target_test, target_pred))
```
