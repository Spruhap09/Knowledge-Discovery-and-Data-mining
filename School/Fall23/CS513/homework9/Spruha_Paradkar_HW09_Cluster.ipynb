```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print('Loading the preprocessed data')
data = pd.read_csv('/Users/spruhap/Desktop/School/Fall23/CS513/wisc_bc_ContinuousVar.csv')
data.head(10)
```

```python
print('The length of the database before dropping all the missing values')
print(len(data))
```

```python
# finding the missing values
print('Finding the missing values')
print(data.isnull().sum())
```

```python
# find '?' values in entire dataset
print('Finding the ? values in the entire dataset')
print(data[data == '?'].count())
```

```python
print('Dropping the id and diagnosis fields')
# Splitting the dataset
attributes = data.drop(columns = ['id', 'diagnosis'])
attributes.head(10)
```

```python
print('Creating a data frame called target for the Class variable')
target = data['diagnosis'] # target variable
target.head(10)
```

```python
# scale the data
scaler = StandardScaler()

# Fit and transform the data
attributes = pd.DataFrame(scaler.fit_transform(attributes), columns=attributes.columns)
attributes.head()
```

```python

print('Dividing the data into test and train data given the 30/70 split ratio')
attribute_train, attribute_test, target_train,target_test  = train_test_split(attributes, target,test_size = 0.3, random_state = 7, shuffle = True)
```

```python
print('doing kmeans')
num_clusters = 2  # dataset has 4 discharge location values (0-3)
kmeans = KMeans(n_clusters=num_clusters, random_state=12)
kmeans.fit(attribute_train)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
df_cluster = pd.DataFrame({'Actual': target_train, 'Cluster': labels})
# Create a cross-tabulation
cross_tab = pd.crosstab(df_cluster['Actual'], df_cluster['Cluster'])

print(cross_tab)
```

```python
preds = kmeans.predict(attribute_test)

# Create a cross-tabulation
cross_tab = pd.crosstab(target_test, preds)

print(cross_tab)
```

```python
# map B = 1 and M = 0
target_test = target_test.map({'B': 1, 'M': 0})

print(f"Accuracy= {accuracy_score(target_test, preds)}")
print()
cm=confusion_matrix(target_test, preds)
print(cm)
plt.figure(figsize=(10, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap=plt.cm.Reds)
plt.show()
print()
print(classification_report(target_test,preds))
```

```python
print('doing the same for hclust')
from sklearn.cluster import AgglomerativeClustering

groups = AgglomerativeClustering(n_clusters=2, metric='euclidean',linkage='ward')

#clusters=groups.fit_predict(attributes)
clusters = groups.fit(attribute_train)
```

```python
import pandas as pd
df_cluster=pd.DataFrame({'Actual':target,'Cluster':clusters})
# Create a cross-tabulation
cross_tab = pd.crosstab(df_cluster['Actual'], df_cluster['Cluster'])

print(cross_tab)
```

```python
preds = groups.fit_predict(attribute_test)
```

```python
print(f"Accuracy= {accuracy_score(target_test, preds)}")
print()
cm=confusion_matrix(target_test, preds)
print(cm)
plt.figure(figsize=(10, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap=plt.cm.Reds)
plt.show()
print()
print(classification_report(target_test,preds))
```
