```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

breastCancer = pd.read_csv('/Users/spruhap/Downloads/breast-cancer-wisconsin.csv')
breastCancer.head(10)
```

```python
print('Summary of F6')
print()

print('getting the missing values for F6')
breastCancer2 = breastCancer.loc[breastCancer['F6'] == '?']
print('The number of columns with missing values are:')
print(len(breastCancer2))
breastCancer2.head(10)
```

```python
breastCancer3 = breastCancer.loc[breastCancer['F6'] != '?']
mean = breastCancer3['F6'].astype(int).mean()
print('The mean of F6 after removing the missing values is')
print(mean)
print()

breastCancer["F6"].replace("?", str(mean), inplace=True)
breastCancer['F6'] = breastCancer["F6"].astype(float)
print('Summary of F6 after replacing the missing values with the mean')
print(breastCancer['F6'].describe())
```

```python
print('Displaying the frequency table of Class vs. F6')
print()
breastCancer['F6'] = breastCancer["F6"].astype(int)
frequencyTable = pd.crosstab(breastCancer['Class'], breastCancer['F6']).style.background_gradient(cmap='YlGn')
frequencyTable
```

```python
import seaborn as sns
print('Displaying the scatter plot of F1 to F6, one pair at a time')
print()
columns = breastCancer.drop(columns = ['Class', 'Sample']);
sns.pairplot(columns)
plt.show()
```
